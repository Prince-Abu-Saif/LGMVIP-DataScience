###**Task 2 : Image to Pencil sketch Conversion**
Image: [Iron Man](https://drive.google.com/file/d/1Ck7_9xGCEhWZ42LtmPSEeMchSLhR11LJ/view?usp=sharing)

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Iron Man.jpg")
cv2_imshow(img)
```

```python
img.shape
```

```python
resized = cv2.resize(img,(622,350),3)
cv2_imshow(resized)
```

```python
gray = cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
inverted_gray = 255-gray
cv2_imshow(inverted_gray)
```

```python
blur = cv2.GaussianBlur(inverted_gray,(21,21),0)
cv2_imshow(blur)
```

```python
inverted_blur = 255-blur 
cv2_imshow(inverted_blur)
```

```python
sketch = cv2.divide(gray,inverted_blur,scale=256.0)
cv2_imshow(sketch)
```
